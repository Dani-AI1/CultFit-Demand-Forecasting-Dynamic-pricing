```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import the dataset:

df=pd.read_csv(r'D:\RegAnalysis - Project\Project 2\Class Apr-May 2018.csv')
df1=pd.read_csv(r'D:\RegAnalysis - Project\Project 2\Class june 2018.csv')
```

```python
# Checking on the various features with our Target variable (Numbers booked)

df.head(20)
df.shape

df1.head(20)
df1.shape
```

```python
# Since both datasets have similar colums we can stack them using concat

df2=pd.concat([df, df1], ignore_index=True)
```

```python
df2.head()
df2.shape
```

```python
# Data Cleaning and Pre-processing:

#Handling null values:

df2.isnull().sum()
```

```python
## Impute the Price(INR) using median:

df2['Price (INR)']= df2['Price (INR)'].fillna(df2['Price (INR)'].median())

```

```python
df2.isnull().sum()
```

```python
# Obeserving the datatypes of the variables:
df2.dtypes
```

```python
df2['ActivitySiteID'].value_counts().reset_index
```

```python
# Combining the columns BookingEndDateTime (Month / Day / Year) & BookingStartTime to Change the datatype into .datetime() format:

combined_string = df2['BookingEndDateTime (Month / Day / Year)']+' '+df2['BookingStartTime']
combined_string
```

```python
df2['Full_datetime']=pd.to_datetime(combined_string, format='mixed')
df2.head()
```

```python
# Checking on changed column:

df2.dtypes
```

```python
## Finding the duplicate values:

df2.duplicated().sum()
```

```python
df2.isnull().sum().index
```

```python
## Feature Engineering:

df2['Hour']=df2['Full_datetime'].dt.hour
df2['DayOfWeek']=df2['Full_datetime'].dt.day_name()
```

```python
print(df2[['Full_datetime', 'Hour', 'DayOfWeek']].head())
```

```python
df2.head()
```

```python
df2.dtypes
```

```python
df2['Hour'] = df2['Hour'].astype(str)
```

```python
# Encoding the following columns:

df2_dummies = pd.get_dummies(df2, columns=['ActivitySiteID', 'DayOfWeek', 'Hour'], dtype=int, drop_first=True)
df2_dummies.head()
```

```python
# Feature selection:

x = df2_dummies.drop(columns=['Number Booked','ActivityDescription','BookingEndDateTime (Month / Day / Year)','BookingStartTime','Full_datetime'])
y = df2_dummies['Number Booked']
```

```python
# To split the train and test data:

from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
print('x_train',x_train.shape)
print('x_test',x_test.shape)
print('y_train',y_train.shape)
print('y_test',y_test.shape)
```
